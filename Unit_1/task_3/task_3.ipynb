Для hh.ru:

1. Сгруппируйте вакансии по направлениями (DS, DE, Software Engenering, etc.)
2. Какая средняя и медианная зарплата по группам вакансий?
3. Какая средняя и медианная зарплата по каждому региону?
4. Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?
5. Какое процентное соотношение каждого региона по вакансиям от всех вакансий?
6. Какая корреляция уровня опыта от зарплаты? (Преобразовать уровень опыта в числовой тип, если возможно. Например: 'MI': 0, 'SE': 1, 'EN': 2, 'EX': 3)
7. Сколько должностей в наборе данных?
8. Какие 10 наиболее часто встречающихся должностей?

```python
data = pd.read_csv(Path().cwd() / 'ds_salaries.csv',  index_col=0)
```

#### 1. Сгруппируйте вакансии по направлениями (DS, DE, Software Engenering, etc.)

```python
gr = data.groupby('job_title')
```

#### 2. Какая средняя и медианная зарплата по группам вакансий?

```python
gr.salary_in_usd.agg(['mean', 'median']).round(2)
```

#### 3. Какая средняя и медианная зарплата по каждому региону?

```python
data.groupby('company_location').salary_in_usd.agg(['mean', 'median']).round(2)
```

#### 5. Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?

```python
gr.salary_in_usd.agg(['mean', 'median']).round(2).nlargest(1, 'mean')
```

#### 6. Какое процентное соотношение каждого региона по вакансиям от всех вакансий?

```python
data['company_location'].value_counts()/data['company_location'].count()
```

#### 7. Какая корреляция уровня опыта от зарплаты? (Преобразовать уровень опыта в числовой тип, если возможно. Например: 'MI': 0, 'SE': 1, 'EN': 2, 'EX': 3)

```python
data['experience_level'] = data['experience_level'].replace({'MI': 0, 'SE': 1, 'EN': 2, 'EX': 3})
```

```python
data[['experience_level', 'salary_in_usd']].corr()
```

#### 8. Сколько должностей в наборе данных?

```python
data['job_title'].unique().size
```

#### 9. Какие 10 наиболее часто встречающихся должностей?

```python
data['job_title'].value_counts()[:10]
```
